```python
t = np.linspace(0, 1000, 1000, endpoint=False)
A = 1
sinusoid = A*(np.sin(t))*(np.sin(sqrt(2)*t))
noise = np.random.rand(1000)-0.5
signal=sinusoid+0*noise
signal1=sinusoid+1*noise
```
