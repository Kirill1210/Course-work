```python
import matplotlib.pyplot as plt
import decimal
import random
import pandas as pd
import seaborn as sns
import matplotlib.ticker as ticker
import numpy as np
import math
import numpy
from scipy import signal
import seaborn as sns
from matplotlib.pyplot import figure
from math import sqrt
 
```

```python
t = np.linspace(0, 1000, 1000, endpoint=False)
A = 1
sinusoid = A*(np.sin(t))*(np.sin(sqrt(2)*t))
noise = np.random.rand(1000)-0.5
signal=sinusoid+0*noise
signal1=sinusoid+1*noise
```

```python
figure(figsize=(8, 6), dpi=80)
plt.plot(t,sinusoid)
plt.title('График сигнала')
plt.ylabel('Частота')
plt.xlabel('t, cекунд')
plt.grid()
plt.xlim(0, 1000)
```

```python
figure(figsize=(8, 6), dpi=80)
plt.plot(t,noise)
plt.title('График шума')
plt.ylabel('Частота')
plt.xlabel('t, cекунд')
plt.grid()
plt.xlim(0, 1000)
```

```python
figure(figsize=(8, 6), dpi=80)
plt.plot(t,signal)
plt.title('График сигнала c шумом при λ=0 ')
plt.ylabel('Частота')
plt.xlabel('t, cекунд')
plt.grid()
plt.xlim(0, 1000)
```
